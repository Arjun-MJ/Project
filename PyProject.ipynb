```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset into pandas dataframe
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df=pd.read_csv(url)
```

```python
#exploring data
print(df.head())  #it will display the first few rows
print(df.info())  #it will display the datatypes and non null counts
print(df.describe())  #summary statistics for numerical columns
```

```python
#setting aesthetic style of the plots
sns.set(style="whitegrid")
```

```python
# 1. count plot for survival
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Survived')
plt.title('count of survival')
plt.xlabel('Survived (0=no,1=yes)')
plt.ylabel('count')
plt.show()

```

```python
# 2. bar plot for survival by gender
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Sex', y='Survived', errorbar=None)
plt.title('survival rate by gender')
plt.xlabel('gender')
plt.ylabel('survival rate')
plt.show()
```

```python
#3.histogram of age distribution 
plt.figure(figsize=(10,6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of passengers')
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```

```python
# 4. Box plot of fare by survival status
plt.figure(figsize=(10, 6))
sns.boxenplot(data=df, x='Survived', y='Fare')
plt.title('fare Distribution by survival status')
plt.xlabel('survived (0= NO ,1 =YES)')
plt.ylabel('Fare')
plt.show()
```
